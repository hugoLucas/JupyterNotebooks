SVMs are particularly well suited for classification of complicated small/medium databases.

# Linear SVM Classification

An SVM classifier is the widest possible street between classes (**large margin classification**). Thus adding more training instances off the street will not impact the decision boundary as this boundary is only determined by the instances located on the "edge"(these instances are called **support vectors**). 

SVMs are sensitive to feature scaling. Without scaling the width of the street may come out much smaller. 

### Softmargin Classification

**Hard Margin Classification** is the requirment that all be off the street and on the right side. Only works if the data is linearly seperable and if the data has no outliers. 

**Soft Margin Classification** is the balance between keeping the street as wide as possible and limiting amount of instances that end up on the wrong side of the street or in the middle of the street (**margin violations**).

For SVMs made in Scikit-learn, this balance is controlled by the **C** parameter. If the parameter is high, you get a smaller margin with fewer margin violations. The converse is true for high values. If an SVM model is overfitting, try reducing C.

```python
# Code shows how to train an SVM model using Scikit-learn

import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.float64)  # Iris-Virginica

scaler = StandardScaler()
svm_clf1 = LinearSVC(C=100, loss="hinge")
svm_clf2 = LinearSVC(C=1, loss="hinge")

scaled_svm_clf1 = Pipeline((
        ("scaler", scaler),
        ("linear_svc", svm_clf1),
    ))
scaled_svm_clf2 = Pipeline((
        ("scaler", scaler),
        ("linear_svc", svm_clf2),
    ))

scaled_svm_clf1.fit(X, y)
scaled_svm_clf2.fit(X, y)
```

Using the SVC class with kernel parameter 'linear' is slower. You can also use a SGD model with parameters $loss=hinge$ and $alpha=1/(m*C)$ to train an SVM. Use it to handle large datasets and online classification tasks.

# Nonlinear SVM Classification

One approach to handle non-linear data is to add polynomial features. 

```python
X, y = make_moons(n_samples=100, noise=0.15, random_state=42)

def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
# Applies a polynomial feature transfomer, a scaler, and finally trains an SVM on the moons database
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

polynomial_svm_clf = Pipeline((
        ("poly_features", PolynomialFeatures(degree=3)),
        ("scaler", StandardScaler()),
        ("svm_clf", LinearSVC(C=10, loss="hinge"))
    ))

polynomial_svm_clf.fit(X, y)

# Plots the results 
def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
    
def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

plot_predictions(polynomial_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])

plt.show()
```

### Polynomial Kernel

Adding polynomial features is too slow at large degrees and insufficient to describe complex data at small degrees. You can use a **kernel trick** instead in order to avoid the exponential explosion of features when handling complex data.

```python
# Application of kernel trick with polynomial of degree 3 and 100
from sklearn.svm import SVC

poly_kernel_svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
    ))

poly100_kernel_svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=10, coef0=100, C=5))
    ))

poly_kernel_svm_clf.fit(X, y)
poly100_kernel_svm_clf.fit(X, y)

plt.figure(figsize=(11, 4))

plt.subplot(121)
plot_predictions(poly_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.title(r"$d=3, r=1, C=5$", fontsize=18)

plt.subplot(122)
plot_predictions(poly100_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.title(r"$d=10, r=100, C=5$", fontsize=18)

plt.show()
```

### Adding Similarity Features

You can also add features computed with a **simularity function** that measures how each instance resembles a particular **landmark**. 

```python
def gaussian_rbf(x, landmark, gamma):
    return np.exp(-gamma * np.linalg.norm(x - landmark, axis=1)**2)

gamma = 0.3

x1s = np.linspace(-4.5, 4.5, 200).reshape(-1, 1)
x2s = gaussian_rbf(x1s, -2, gamma)
x3s = gaussian_rbf(x1s, 1, gamma)

X1D = np.linspace(-4, 4, 9).reshape(-1, 1)
XK = np.c_[gaussian_rbf(X1D, -2, gamma), gaussian_rbf(X1D, 1, gamma)]
yk = np.array([0, 0, 1, 1, 1, 1, 1, 0, 0])

plt.figure(figsize=(11, 4))

plt.subplot(121)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.scatter(x=[-2, 1], y=[0, 0], s=150, alpha=0.5, c="red")
plt.plot(X1D[:, 0][yk==0], np.zeros(4), "bs")
plt.plot(X1D[:, 0][yk==1], np.zeros(5), "g^")
plt.plot(x1s, x2s, "g--")
plt.plot(x1s, x3s, "b:")
plt.gca().get_yaxis().set_ticks([0, 0.25, 0.5, 0.75, 1])
plt.xlabel(r"$x_1$", fontsize=20)
plt.ylabel(r"Similarity", fontsize=14)
plt.annotate(r'$\mathbf{x}$',
             xy=(X1D[3, 0], 0),
             xytext=(-0.5, 0.20),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18,
            )
plt.text(-2, 0.9, "$x_2$", ha="center", fontsize=20)
plt.text(1, 0.9, "$x_3$", ha="center", fontsize=20)
plt.axis([-4.5, 4.5, -0.1, 1.1])

plt.subplot(122)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(XK[:, 0][yk==0], XK[:, 1][yk==0], "bs")
plt.plot(XK[:, 0][yk==1], XK[:, 1][yk==1], "g^")
plt.xlabel(r"$x_2$", fontsize=20)
plt.ylabel(r"$x_3$  ", fontsize=20, rotation=0)
plt.annotate(r'$\phi\left(\mathbf{x}\right)$',
             xy=(XK[3, 0], XK[3, 1]),
             xytext=(0.65, 0.50),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18,
            )
plt.plot([-0.1, 1.1], [0.57, -0.1], "r--", linewidth=3)
plt.axis([-0.1, 1.1, -0.1, 1.1])
    
plt.subplots_adjust(right=1)

plt.show()
```

In the left graph, two landmarks at $x_1 = -2$ and $x_1 = 1$ are added. Here we define the simularity function to be the **Gaussian Radial Bias Function (RBF)** with $\gamma = 0.3$: $\phi \gamma (x,\ell) = exp(-\gamma \, ||x - \ell||^2)$. You use this function to compute new features and drop the old ones (Ex. $x_1$ at -1 is located 1 from the first landmark and 2 from the other thus we add the features $x_2 = exp(-0.3 \times 1^2) \approx 0.74$ and $x_3 = exp(-0.3 \times 2^2) \approx 0.30$). A naive way to select landmarks would be to put a landmark at each an every instance. However this results in a training set of $m$ features and $m$ instances when you drop all the old features.

### Gaussian RBF Kernel

```python
# Example of SVC class using Gaussian RBF 
rbf_kernel_svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
    ))
rbf_kernel_svm_clf.fit(X, y)
```

```python
from sklearn.svm import SVC

gamma1, gamma2 = 0.1, 5
C1, C2 = 0.001, 1000
hyperparams = (gamma1, C1), (gamma1, C2), (gamma2, C1), (gamma2, C2)

svm_clfs = []
for gamma, C in hyperparams:
    rbf_kernel_svm_clf = Pipeline((
            ("scaler", StandardScaler()),
            ("svm_clf", SVC(kernel="rbf", gamma=gamma, C=C))
        ))
    rbf_kernel_svm_clf.fit(X, y)
    svm_clfs.append(rbf_kernel_svm_clf)

plt.figure(figsize=(20, 15))

for i, svm_clf in enumerate(svm_clfs):
    plt.subplot(221 + i)
    plot_predictions(svm_clf, [-1.5, 2.5, -1, 1.5])
    plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
    gamma, C = hyperparams[i]
    plt.title(r"$\gamma = {}, C = {}$".format(gamma, C), fontsize=16)

plt.show()
```

The parameter $\gamma$ acts as a regularization hyperparameter. If you're overfitting, reduce it. If you're undefitting, increase it. Increasing $\gamma$ makes the bell-shaped curve narrower. 

As a rule of thumb, always try the linear kernel first. If the training set is not too large, try the Gaussian RBF as well. Only when you have spare time, or when you know a specific kernel is tailored to your problem domain, should you experiment with other kernels.

### Computation Complexity

The LinearSVC class does not support the *kernel trick* mention above so it scales roughly to $O(n \times m)$. It takes longer if the tolerance, $\epsilon$, is set to a very high percision. 

The SVC supports the *kernel trick* but its complexity ranges from $O(m^2 \times n)$ to $O(m^3 \times n)$. Thus it should be used when features are sparse or when the data set is small/medium sized and there are lots of features. 

# SVM Regression

**SVM Regression** tries to fit as many instances as possible *on* the street while limiting the margin violations. Width of the street is controlled by parameter $\epsilon$. 

```python
from sklearn.svm import LinearSVR
import numpy.random as rnd

rnd.seed(42)
m = 50
X = 2 * rnd.rand(m, 1)
y = (4 + 3 * X + rnd.randn(m, 1)).ravel()

svm_reg = LinearSVR(epsilon=1.5)
svm_reg.fit(X, y)

svm_reg1 = LinearSVR(epsilon=1.5)
svm_reg2 = LinearSVR(epsilon=0.5)
svm_reg1.fit(X, y)
svm_reg2.fit(X, y)

def find_support_vectors(svm_reg, X, y):
    y_pred = svm_reg.predict(X)
    off_margin = (np.abs(y - y_pred) >= svm_reg.epsilon)
    return np.argwhere(off_margin)

svm_reg1.support_ = find_support_vectors(svm_reg1, X, y)
svm_reg2.support_ = find_support_vectors(svm_reg2, X, y)

eps_x1 = 1
eps_y_pred = svm_reg1.predict([[eps_x1]])

def plot_svm_regression(svm_reg, X, y, axes):
    x1s = np.linspace(axes[0], axes[1], 100).reshape(100, 1)
    y_pred = svm_reg.predict(x1s)
    plt.plot(x1s, y_pred, "k-", linewidth=2, label=r"$\hat{y}$")
    plt.plot(x1s, y_pred + svm_reg.epsilon, "k--")
    plt.plot(x1s, y_pred - svm_reg.epsilon, "k--")
    plt.scatter(X[svm_reg.support_], y[svm_reg.support_], s=180, facecolors='#FFAAAA')
    plt.plot(X, y, "bo")
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.legend(loc="upper left", fontsize=18)
    plt.axis(axes)

plt.figure(figsize=(9, 4))
plt.subplot(121)
plot_svm_regression(svm_reg1, X, y, [0, 2, 3, 11])
plt.title(r"$\epsilon = {}$".format(svm_reg1.epsilon), fontsize=18)
plt.ylabel(r"$y$", fontsize=18, rotation=0)
plt.plot([eps_x1, eps_x1], [eps_y_pred, eps_y_pred - svm_reg1.epsilon], "k-", linewidth=2)
plt.annotate(
        '', xy=(eps_x1, eps_y_pred), xycoords='data',
        xytext=(eps_x1, eps_y_pred - svm_reg1.epsilon),
        textcoords='data', arrowprops={'arrowstyle': '<->', 'linewidth': 1.5}
    )
plt.text(0.91, 5.6, r"$\epsilon$", fontsize=20)
plt.subplot(122)
plot_svm_regression(svm_reg2, X, y, [0, 2, 3, 11])
plt.title(r"$\epsilon = {}$".format(svm_reg2.epsilon), fontsize=18)
plt.show()
```

Plots above show the difference a small and large $\epsilon$ can make with SVM regression. Adding more training instances within the margin will not change the model's predictions. This is called **$\epsilon$-insensitive**.

```python
from sklearn.svm import SVR

rnd.seed(42)
m = 100
X = 2 * rnd.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + rnd.randn(m, 1)/10).ravel()

svm_poly_reg = SVR(kernel="poly", degree=2, C=100, epsilon=0.1)
svm_poly_reg.fit(X, y)

svm_poly_reg1 = SVR(kernel="poly", degree=2, C=100, epsilon=0.1)
svm_poly_reg2 = SVR(kernel="poly", degree=2, C=0.01, epsilon=0.1)
svm_poly_reg1.fit(X, y)
svm_poly_reg2.fit(X, y)

plt.figure(figsize=(9, 4))
plt.subplot(121)
plot_svm_regression(svm_poly_reg1, X, y, [-1, 1, 0, 1])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg1.degree, svm_poly_reg1.C, svm_poly_reg1.epsilon), fontsize=18)
plt.ylabel(r"$y$", fontsize=18, rotation=0)
plt.subplot(122)
plot_svm_regression(svm_poly_reg2, X, y, [-1, 1, 0, 1])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg2.degree, svm_poly_reg2.C, svm_poly_reg2.epsilon), fontsize=18)
plt.show()
```

To takle nonlinear regression you must use a kernalized SVM model. Figures above show a randomized quadratic set using a $2^{nd}$-degree polynomial kernel. There is more regularization on the left model (small C-value). The above uses the *SVR* class which is the regression equivalent of the *SVC* class. 

# Under the Hood

In the following sections the bias term is denoted by $b$ and the feature weights vector by $w$. Furthermore, no bias feature will be added to the input features vector

### Decision Function and Predictions

Predictions on $x$ from decision function: $w^{T} \cdot x + b = w_1 x_1 + \, ... \, + w_n x_n + b$. The result, $\hat{y}$, is 1 if the sum is positive and 0 if it is negative. The decision boundary is the set of points where the decision function is equal to 0. The margin width is bounded by the set of points where the decision function is equal to 1 and where it is equal to -1. Training an SVM classifier means finding the value of $w$ and b that makes the margin as wide as possible while avoiding margin violations (hard margin) or limiting them (soft margin).

### Training Objective

The slope of the decision function is equal to the norm of the weight vector: $|| \,w \,||$. Dividing the slope by $n$ makes the margin $n$ time wider. Need to minimize $|| \,w \,||$ to get large margin. 

For a hard margin we need the decision function to be greater than 1 for positive instances and less than -1 for negative instances: $t^{(i)}\,(\,w^{T} \cdot x^{(i)} + b \,) \geq 1$ where $t^{(i)} = -1$ when $y^{(i)} = 0$ and  $t^{(i)} = 1$ when $y^{(i)} = 1$.

The hard margin SVM classifier objective is the *constrained optimization* problem: 
$minimize_{w,b} \,\,\, \frac{1}{2}w^T \cdot w$
subject to $\,\, t^{(i)}\,(\,w^{T} \cdot x^{(i)} + b \,) \geq 1$

The slack variable $\varsigma^{(i)}$ (which must be greater than or equal to zero for all instances) is how much each instance is allowed to violate the margin constraint. The soft margin SVM classifier objective: 
$minimize_{w,b,\varsigma} \,\,\, \frac{1}{2}w^T \cdot w + C \Sigma_{i=1}^{m} \varsigma^{(i)}$ 
subject to $\,\, t^{(i)}\,(\,w^{T} \cdot x^{(i)} + b \,) \geq 1 - \varsigma^{(i)}$ and $\varsigma^{(i)} \geq 0$ for $i = 1, 2, \, ... \, , m$

### Quadratic Programming

The hard and soft margin problems are both convex quadratic optimization problems with linear constraints a.k.a *Quadratic Programming Problems*. 

### The Dual Problem

Given a constrained optimization problem, its possible to express a different but closely related problem called its **dual problem**. The solution to the dual problem in general typically gives the lower bound to the solution of the constrained optimization problem, for an SVM it can result in the solutions being equal however. 

Dual form of the SVM problem: $minimize_{\alpha} \frac{1}{2} \, \Sigma_{i=1}^m \, \Sigma_{j=1}^m \, \alpha^{(i)} \alpha^{(j)} \, t^{(i)} t^{(j)} \, x^{(i)^T} \cdot x^{(j)} - \, \Sigma_{i=1}^m \alpha^{(i)} $ subject to $\alpha^{(i)} \geq 0$ for $i = 1,2, \, ... \, , m$. A QP solver must be used to minimize $\alpha$ which is then fed to Equation 5-7 in the book to calculate the weight vector and bias terms. The dual problem is faster to solve than the constrained optimization problem when the number of training instances is smaller than the amount of features.  It also makes the kernel trick possible. 

### Kernelized SVM

The second degree polynomial mapping function is given by $\phi(x)$ and is a $1 \times 3$ vector. Applying it to two vectors $a$ and $b$ and computing their dot poduct results in: $\phi(a)^T \cdot \phi(b) = \, .... \, = (a^T \cdot b)^2$. Applying this transformation to all training instances makes solving the dual problem formulation of the constrained optimization problem easier as the dot product term in Equation 5-6 can be replaced by the square of the dot product. The end result is that you don't actually need to do the transformation but you can still replace the term in Equation 5-6. This results in a more computaionally efficient solution. This is the kernel trick. 

### Online SVMs

One method of doing an online SVM is applying Gradient Descent on a cost function derived from the primal (constrained optimization) problem.
