```python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(clf, X, y, axes=[-1.5, 2.5, -1, 1.5], alpha=0.5, contour=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap, linewidth=10)
    if contour:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", alpha=alpha)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", alpha=alpha)
    plt.axis(axes)
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
```

A group of predictors is called an **ensemble**. Using an ensemble of Decision Trees is called a **Random Forrest**. 

# Voting Classifiers

In a **hard voting classifier** (HVC), you aggregate the predictions of multiple models and predict the prediction with the most votes (i.e. the predicition that most of our models output). Even if each ensemble model is a **weak learner** (makes predictions no better than chance) your HVC can be a **strong learner**  (makes predictions with high accuracy) if provided enough diverse weak learners. This is a result of the *law of large numbers*. Keep in mind that that each model should make predictions independant of the other models in order to make sure they all make different kinds of errors.

The following code trains an HVC composed of three classifers on the moons database: 

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

log_clf = LogisticRegression(random_state=42)
rnd_clf = RandomForestClassifier(random_state=42)
svm_clf = SVC(random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard')
voting_clf.fit(X_train, y_train)
```

Now let's evaluate the accuracy of the individual classifiers and of the HVC:

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

The HVC performs slightly better than the individual models. 

Another metric is called **soft voting** in which the prediction of the ensemble model is the probability class with the highest average probability amongst all the components. To use this in Scikit-Learn, the classifier must have a $dic\_proba()$ method and you must set the $voting$ parameter to "soft". A soft voting classifier (SoVC) usually performs better than an HVC.  

# Bagging and Pasting

Another way to get a diverse set of predictors is to train the same algorithm on different random subsets of the training data. When sample is done with repalcement this method is called **bagging**, when done without replacment it's called **pasting**. Both allow a particular instance to be used by different predictors, but only bagging allows a predictor to use the same instance multiple times. 

End result is that ensemble has a similar bias but a lower variance than a single predictor trained on the original training set. 

### Bagging and Pasting in Scikit-Learn

In Scikit-Learn the $BaggingClassifier$ class is used for bagging and pasting. The following code snipet is an example of bagging using 500 Decision Trees each training on 100 training instances.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Automatically uses soft voting iff the base clasifier has a predict_proba() method
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1, random_state=42)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
y_pred_tree = tree_clf.predict(X_test)
print(accuracy_score(y_test, y_pred_tree))
```

Compared to a single Decision Tree, the bagging method results in an increased accuracy. 

```python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(clf, X, y, axes=[-1.5, 2.5, -1, 1.5], alpha=0.5, contour=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap, linewidth=10)
    if contour:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", alpha=alpha)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", alpha=alpha)
    plt.axis(axes)
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
```

```python
plt.figure(figsize=(11,4))
plt.subplot(121)
plot_decision_boundary(tree_clf, X, y)
plt.title("Decision Tree", fontsize=14)
plt.subplot(122)
plot_decision_boundary(bag_clf, X, y)
plt.title("Decision Trees with Bagging", fontsize=14)
plt.show()
```

Compared to a single Decision Tree, the bagging method results in a much more regular boundary. However, the ensemble has roughly the same bias but a lower variance. In general, use bagging. If you have the time though, try bagging and pasting to see which is better. 

### Out-of-Bag Evaluation

On average, only 63% of all instances are sampled for each predictor using bagging. This means 37% of training instances are not sample (i.e. **out-of-the-bag** (oob)). Note that the 37% are different for each predictor.

Since a predictor never sees the oob instances, it can be cross validated on them. Setting the $oob\_scoring$ hyperparameter of the $BaggingClassifier$ to "true" will do this for you automatically. 

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(splitter="random", max_leaf_nodes=16, random_state=42),
    n_estimators=500, max_samples=1.0, bootstrap=True, n_jobs=-1, random_state=42)

bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

# Random Patches and Random Subspaces

Sampling both training instances and features is called the **Random Patches** method. Keeping all training instances but sampling features is called the **random subspaces** method. Sampling features results in even more predictor diversity at the expense of a slightly higher bias for a lower variance.

# Random Forrests

A Random Forrest is an ensemble of Decision Trees using the bagging method with $max\_samples = m$. Scikit has a class named **RandomForrestClassifier** that trains a random forrest for you.

```python
from sklearn.ensemble import RandomForestClassifier

rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)
rnd_clf.fit(X_train, y_train)

y_pred_rf = rnd_clf.predict(X_test)
```

The random forrest algorithm looks for the best feature amongst a subset of features which results in higher tree diversity. 

### Extra Trees

An **Extremely Randomized Tree** or **Extra Tree** for short, is a tree that uses a random threshold for splitting rather than taking the time to find the most optimum threshold.  This makes the extra tree very fast to train but may not result in a more accurate model when compared to the random forrest method. The only way to tell is to try both. 

### Feature Importance

It is possible to estimate the importance of a feature by computing the average depth of the feature for each decision tree in a random forrest. Scikit-learn does this automatically and stores the result in the **feature\_importances\_** variable. The following is an example of this on the Iris dataset.

```python
from sklearn.datasets import load_iris
iris = load_iris()
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=42)
rnd_clf.fit(iris["data"], iris["target"])
for name, score in zip(iris["feature_names"], rnd_clf.feature_importances_):
    print(name, score)
```

Thus the most important features are the petal length and petal width. 

# Boosting

**Boosting** (aka **Hypothesis Boosting**) refers to any ensemble method that can combine serveral weak learners into a strong learner. General idea is to train a sequence of learners with each new learner correcting the mistakes of its predacessor. 

### Ada Boost

In **Adaptive Boosting** (AdaBoost) each learner in a seqence pays extra attention to the training instances its predacessor got wrong. For example one could start with training a Decision Tree and then add extra weight to the training instances the tree misclassified. The next tree would then be trained and focus on those examples a bit more. 

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=200,
    algorithm="SAMME.R", learning_rate=0.5, random_state=42)
ada_clf.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC

m = len(X_train)

plt.figure(figsize=(11, 4))
for subplot, learning_rate in ((121, 1), (122, 0.5)):
    sample_weights = np.ones(m)
    for i in range(5):
        plt.subplot(subplot)
        svm_clf = SVC(kernel="rbf", C=0.05, random_state=42)
        svm_clf.fit(X_train, y_train, sample_weight=sample_weights)
        y_pred = svm_clf.predict(X_train)
        sample_weights[y_pred != y_train] *= (1 + learning_rate)
        plot_decision_boundary(svm_clf, X, y, alpha=0.2)
        plt.title("learning_rate = {}".format(learning_rate - 1), fontsize=16)

plt.subplot(121)
plt.text(-0.7, -0.65, "1", fontsize=14)
plt.text(-0.6, -0.10, "2", fontsize=14)
plt.text(-0.5,  0.10, "3", fontsize=14)
plt.text(-0.4,  0.55, "4", fontsize=14)
plt.text(-0.3,  0.90, "5", fontsize=14)
plt.show()
```

As you can see, AdaBoosting resembles gradient descent in that the method is iterative. However, instead of tweeking a parameter to minimize the cost function like DG, AdaBoost just adds more learners. **Note:** due to the sequential nature of AdaBoosting, it is not possible to parallelize the entire method. Therefore it does not scale as well as bagging or pasting. 

At the start of the algorithm, each instance is assigned a weight $w^{(i)}$ of $\frac{1}{m}$. The algorithm is then trained at the error rate is calculated: $ \large r_j =  \frac{\Sigma_{i=1 \, \& \, \hat{y}^{(i)}_j \neq y^{(i)}}^{m} \, w^{(i)}}{\Sigma_{i=1}^{m} \, w^{(i)}}$ where $\hat{y}^{(i)}_j$ is the $j^{th}$ predictor's predictor for the i^{th} instance. The predictor's weight is then computed by $\alpha = \eta \, log \frac{1 - r_j}{r_j}$ where $\eta$ is the learning rate. Next, for all weights, the weight is unchanged if the prediction was correct and is set to $w^{(i)} exp(a_j)$ if the prediction was wrong. The weights are then normalized. 

### Gradient Boosting

Like AdaBoosting, **Gradient Boosting** trains a sequence of predictors. Unlike it however, gradient boosting trains the new predictor on the *residual error* made by the previous predictors. First we train a Decision Tree to have a base:

```python
from sklearn.tree import DecisionTreeRegressor

np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)

tree_reg1 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg1.fit(X, y)
```

Now we train a second on the residual errors of the first:

```python
y2 = y - tree_reg1.predict(X)
tree_reg2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg2.fit(X, y2)
```

And then a third:

```python
y3 = y2 - tree_reg2.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg3.fit(X, y3)
```

And now we make a prediction by adding the results of each component:

```python
X_new = np.array([[0.8]])
y_pred = sum(tree.predict(X_new) for tree in (tree_reg1, tree_reg2, tree_reg3))
y_pred
```

We can also see it in action in the graphs below:

```python
def plot_predictions(regressors, X, y, axes, label=None, style="r-", data_style="b.", data_label=None):
    x1 = np.linspace(axes[0], axes[1], 500)
    y_pred = sum(regressor.predict(x1.reshape(-1, 1)) for regressor in regressors)
    plt.plot(X[:, 0], y, data_style, label=data_label)
    plt.plot(x1, y_pred, style, linewidth=2, label=label)
    if label or data_label:
        plt.legend(loc="upper center", fontsize=16)
    plt.axis(axes)

plt.figure(figsize=(11,11))

plt.subplot(321)
plot_predictions([tree_reg1], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h_1(x_1)$", style="g-", data_label="Training set")
plt.ylabel("$y$", fontsize=16, rotation=0)
plt.title("Residuals and tree predictions", fontsize=16)

plt.subplot(322)
plot_predictions([tree_reg1], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h(x_1) = h_1(x_1)$", data_label="Training set")
plt.ylabel("$y$", fontsize=16, rotation=0)
plt.title("Ensemble predictions", fontsize=16)

plt.subplot(323)
plot_predictions([tree_reg2], X, y2, axes=[-0.5, 0.5, -0.5, 0.5], label="$h_2(x_1)$", style="g-", data_style="k+", data_label="Residuals")
plt.ylabel("$y - h_1(x_1)$", fontsize=16)

plt.subplot(324)
plot_predictions([tree_reg1, tree_reg2], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h(x_1) = h_1(x_1) + h_2(x_1)$")
plt.ylabel("$y$", fontsize=16, rotation=0)

plt.subplot(325)
plot_predictions([tree_reg3], X, y3, axes=[-0.5, 0.5, -0.5, 0.5], label="$h_3(x_1)$", style="g-", data_style="k+")
plt.ylabel("$y - h_1(x_1) - h_2(x_1)$", fontsize=16)
plt.xlabel("$x_1$", fontsize=16)

plt.subplot(326)
plot_predictions([tree_reg1, tree_reg2, tree_reg3], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h(x_1) = h_1(x_1) + h_2(x_1) + h_3(x_1)$")
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$y$", fontsize=16, rotation=0)
plt.show()
```

The left hand side shows each individual Decision Tree that was trained while the right displays the sum at that iteration. 

Alternatively, you can also just use a scikit library that does this all for you:

```python
from sklearn.ensemble import GradientBoostingRegressor

gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3, learning_rate=1.0, random_state=42)
gbrt.fit(X, y)
```

The learning rate hyperparameter scales the contribution of each tree. If set to a low value it will take more trees to use gradient boosting by the end result will generalize better. This technique  is called **shrinkage**. 

```python
gbrt_slow = GradientBoostingRegressor(max_depth=2, n_estimators=200, learning_rate=0.1, random_state=42)
gbrt_slow.fit(X, y)

plt.figure(figsize=(11,4))

plt.subplot(122)
plot_predictions([gbrt], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="Ensemble predictions")
plt.title("learning_rate={}, n_estimators={}".format(gbrt.learning_rate, gbrt.n_estimators), fontsize=14)

plt.subplot(121)
plot_predictions([gbrt_slow], X, y, axes=[-0.5, 0.5, -0.1, 0.8])
plt.title("learning_rate={}, n_estimators={}".format(gbrt_slow.learning_rate, gbrt_slow.n_estimators), fontsize=14)

plt.show()
```

# Stacking

In **Stacking** (short for stacked generalization) you use another learning model as a *blender* or *meta learner* to apply some aggregating function to the output of your ensemble model. 

First the training set is split into two. The first subset is used to train first layer predictors which then make *clean* predictions on the second subset. These values are then fed into the *blender*/*meta learner* so that it may learn to predict the correct target value given the predictions from the first layer.
